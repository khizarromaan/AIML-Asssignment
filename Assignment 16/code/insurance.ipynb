```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv")
df.head(10)
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
numerical_cols = ['age', 'bmi', 'children', 'expenses']
categorical_cols = ['sex', 'smoker', 'region']

print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

```python
plt.figure()
sns.histplot(df['age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.histplot(df['bmi'])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.histplot(df['children'])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.histplot(df['expenses'])
plt.title("Charges Distribution")
plt.xlabel("Charges")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.countplot(x='sex', data=df)
plt.title("Count of Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.countplot(x='smoker', data=df)
plt.title("Count of Smoker")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.countplot(x='region', data=df)
plt.title("Count of Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
corr = df[['age', 'bmi', 'children', 'expenses']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
average_charges = df['expenses'].mean()
print("Average charges:", average_charges)
```

```python
max_charges = df['expenses'].max()
min_charges = df['expenses'].min()
print("Maximum charges:", max_charges)
print("Minimum charges:", min_charges)
```

```python
smoker_avg = df.groupby('smoker')['expenses'].mean()
print(smoker_avg)
```

```python
plt.figure()
sns.boxplot(x='smoker', y='expenses', hue='sex', data=df)
plt.title("Charges by Smoker and Sex")
plt.xlabel("Smoker")
plt.ylabel("Charges")
plt.show()
```

```python
avg_age = df['age'].mean()
avg_bmi = df['bmi'].mean()
region_counts = df['region'].value_counts()

print("Average age:", avg_age)
print("Average BMI:", avg_bmi)
print(region_counts)

# Summary:
# 1. The average age is around 39 years and average BMI is around 30.
# 2. Smokers have much higher insurance charges than non-smokers.
# 3. The southeast region has the highest number of customers.
# 4. Charges increase with age and BMI, and smoking has the biggest effect on charges.
```
